```python
import healpy as hp
import numpy as np
import matplotlib.pyplot as plt

# Example settings
nside = 1  # Resolution of the map
npix = hp.nside2npix(nside)
resolution = hp.nside2resol(nside, arcmin=True)/60  # Resolution of the map in degrees

original_map = np.arange(npix)  # A map where value = pixel index for clarity

# Define a rotation
rot = hp.Rotator(rot=[45,45], deg=True)
print(rot(np.pi/4,0))
# Get the angular coordinates for each pixel
theta, phi = hp.pix2ang(nside, np.arange(npix))

# Apply the rotation to these coordinates
theta_rot, phi_rot = rot(theta, phi)
#phi_rot = phi_rot+(0.6*np.ones(12))

print(theta)
print(theta_rot)
print(phi)
print(phi_rot)

# Convert the rotated coordinates back to pixel indices
new_indices = hp.ang2pix(nside, theta_rot, phi_rot)

# Create a new map with the pixels rearranged
rotated_map = original_map[new_indices]
print("figure out the new indices",new_indices)

# Visualize the maps
hp.mollview(original_map, title='Original Map', sub=(1, 2, 1), cmap='viridis')
hp.mollview(rotated_map, title='Rotated Map', sub=(1, 2, 2), cmap='viridis')
plt.show()

print(original_map)
print(rotated_map)
```
